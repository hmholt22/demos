```python
import numpy as np
```

```python
#input data
x = np.array([200, 17])
```

```python
def sigmoid(x):
    return 1.0 / (1.0 + np.exp(-x))

#First layer - three nodes
#Node 1
w1_1 = np.array([1,2]) #Assume w and b are these values as determined by some gradient descent loss minimization process - more later
b1_1 = np.array([-1])
z1_1 = np.dot(w1_1, x)+ b1_1
a1_1 = sigmoid(z1_1)

#Node 2
w1_2 = np.array([-3, 4])
b1_2 = np.array([1])
z1_2 = np.dot(w1_2, x)+ b1_2
a1_2 = sigmoid(z1_2)

#Node 3
w1_3 = np.array([5, -6])
b1_3 = np.array([2])
z1_3 = np.dot(w1_3, x)+ b1_3
a1_3 = sigmoid(z1_3)
```

```python
#Second layer - one node - take the a1's and use them as the data for the next layer!
a1 = np.array([a1_1, a1_2, a1_3])

#Node 1
w2_1 = np.array([-7, 8, 9])
b2_1 = np.array([3])
z2_1 = np.dot(w2_1, a1) + b2_1 #Here we use a1 as inputs for the next layer
a2_1 = sigmoid(z2_1)

a2_1 #Since we've fit everything to sigmoid functions this should be a probability
```

```python
#Writing a Dense layer function

def dense_h(a_in, W, b):
    units = W.shape[1] #W should be a matrix of weights from the first layer. Shape should grab the column dimension, i.e. the # of nodes
    a_out = np.zeros(units)
    for j in range(units):
        w = W[:, j] #Pull jth column
        z = np.dot(w, a_in) + b[j]
        a_out[j] = sigmoid(z)
    return a_out

#using example above
W = np.array([[1, -3, 5], 
            [2, 4, -6]])
b= np.array([-1, 1, 2])
a_in = np.array([-2, 4])

dense_h(a_in, W, b)
```
